```python
#RANDOMFORESTREGRESSOR

from sklearn.ensemble import RandomForestRegressor
RF_Regressor20 = RandomForestRegressor(criterion='squared_error', max_features='sqrt', random_state=1)
param_grid20 = {
    'n_estimators': [800],
    'max_depth': [None]
}
grid_search20 = GridSearchCV(estimator=RF_Regressor20, param_grid=param_grid20, scoring='r2', cv=8)
grid_search20.fit(X_train_pca,Y_train.values.ravel())
```

```python
from pandas import Series


best_parameters20 = grid_search20.best_params_
print("Best parameters: ", best_parameters20)

#R SQUARED
best_result20 = grid_search20.best_score_
print("R-squared (best_score): ", best_result20)

#MODIFIED R SQUARED
r20, c20 = X_train_pca.shape
modified_r220 = 1-(1-best_result20)*(4/5*r20-1)/(4/5*r20-c20-1)
print("modified_r2: ", modified_r220)
```

```python
#TESTING THE RANDOMFOREST REGRESSOR MODEL ON THE TESTING SET

from sklearn.metrics import mean_squared_error, r2_score

Y_pred = grid_search20.best_estimator_.predict(X_test_pca)


mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error on testing set:", mse)
print("R-squared on testing set:", r2)
```

```python
#LINEAR REGRESSION

LinearRegression1 = SGDRegressor(random_state=1, penalty=None)
Hparameter1 = {'eta0': [0.0001],
               'max_iter': [10000,50000]}

grid_search1 = GridSearchCV(estimator=LinearRegression1, param_grid=Hparameter1, scoring='r2', cv=5)
grid_search1.fit(X_train_pca,Y_train.values.ravel())

best_parameters = grid_search1.best_params_
print("Best parameters:", best_parameters)
best_result = grid_search1.best_score_
print("Best result:", best_result)
best_model = grid_search1.best_estimator_
print("Intercept β0:", best_model.intercept_)
```
